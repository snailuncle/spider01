```python
import pymongo
import re
import charts

client=pymongo.MongoClient('localhost',27017)
ceshi=client['ceshi']
item_info=ceshi['item_infoY']
```

```python
def get_all_datas(price_min,price_max):
    interval_size=int((price_max-price_min)/10)
    the_price=price_min
    while the_price<=price_max:
        yield [the_price,the_price+interval_size]
        the_price=the_price+interval_size
```

```python
def get_data_within(price_min,price_max,places):
    for place in places:
        place_regex={'$regex':place}
        price_range_posts=[]
        for data in get_all_datas(price_min,price_max):
            a=list(item_info.find({'place':place_regex,'price':{'$gt':data[0],'$lte':data[1]}}))
            each_day_post=len(a) #指定价格区间的二手商品数量
            price_range_posts.append(each_day_post)
        data={
            'name':place,
            'data':price_range_posts,
            'type':'line'
        }
        yield data
            
        

```

```python
for i in get_data_within(0,100000,['海淀','通州','西城','朝阳']):
    print(i)
```

```python
price_max=10000
options={
    'chart':{'zoomType':'xy'},
    'title':{'text':'58同城3月28日北京二手市场商品信息'},
    'subtitle':{'text':'可视化统计图表'},
    'xAxis':{'categories':[i[1] for i in get_all_datas(0,price_max)]},
    'yAxis':{'title':{'text':'数量'}}
}

# import json
# print(json.dumps(options))

series=[i for i in get_data_within(0,price_max,['海淀','通州','西城','朝阳'])]
charts.plot(series,options=options,show='inline')
```
