```python
import requests
import re,os,time,json,sys
import subprocess as sp






#获取当前时间
time_now = int(time.time())
#转换成localtime
time_local = time.localtime(time_now)
#转换成新的时间格式(2016-05-09 18:59:20)
dt = time.strftime(r"%Y/%m/%d{}%H:%M:%S".format(r'%%20'),time_local)
print(dt)
url='http://ip.zdaye.com/img/m_559259ee216681a2.gif?{}'.format(dt)
print(url)




def timestamp_extract(url):
#     url='http://ip.zdaye.com/img/m_4f63daa4571c0b63.gif?2018/4/6%2019:12:58'
    pattern=r'(?<=\?).*'
    dt=re.search(pattern,url).group()
    dt=re.sub(r'[/%]',' ',dt)
    dt=dt[0:-10]+dt[-8:]
    # print(dt)
    # sys.exit()
    #转换成时间数组
    timeArray = time.strptime(dt, "%Y %m %d %H:%M:%S")
    #转换成时间戳
    timestamp = time.mktime(timeArray)

    return int(timestamp)-20
timestamp=timestamp_extract(url)
#运行本脚本至少等待5分钟
#视网络情况和电脑等级而定.
#不加UA不返回正确信息
old_headers='''
Accept: image/webp,image/apng,image/*,*/*;q=0.8
Accept-Encoding: gzip, deflate
Accept-Language: zh-CN,zh;q=0.9
Connection: keep-alive
Host: ip.zdaye.com
Referer: http://ip.zdaye.com/
User-Agent: Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.32 Safari/537.36%s
'''
old_headers=old_headers.strip().format(timestamp)
print(old_headers)
def headers_format(headers):
    # sub(repl, string[, count]) | re.sub(pattern, repl, string[, count]): 
    new_headers=re.sub(r'(.*): (.*)',r'"\1":"\2",',headers)
    new_headers=new_headers.strip()[0:-1]
    new_headers='{'+new_headers+'}'
    new_headers=json.loads(new_headers)
    return new_headers
headers=headers_format(old_headers)









r=requests.get(url,headers=headers)
print(r)

with open('111111111111.gif','wb') as f:
    f.write(r.content)
```
