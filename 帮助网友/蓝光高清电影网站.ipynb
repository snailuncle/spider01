```python
url='http://www.yinfans.com'

import requests
from lxml import etree

headers={
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding':'gzip, deflate',
    'Accept-Language':'zh-CN,zh;q=0.9',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',
    'Upgrade-Insecure-Requests':'1',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.32 Safari/537.36',
}


html=requests.get(url,headers=headers)

html_text=html.text

print(html_text)

selector=etree.HTML(html_text)





```

```python

title_xpath_str="//li[contains(@class,'box')]/div[@class='article']/h2/a/text()"
link_xpath_str="//li[contains(@class,'box')]/div[@class='article']/h2/a/@href"

titles=selector.xpath(title_xpath_str)
links=selector.xpath(link_xpath_str)


```
