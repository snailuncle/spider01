```python
# 爬取菜鸟教程的代码

url='http://www.runoob.com/python/python-exercise-example4.html'

headers={
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding':'gzip, deflate',
    'Accept-Language':'zh-CN,zh;q=0.9',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',   
    'Host':'www.runoob.com',
    'Upgrade-Insecure-Requests':'1',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.32 Safari/537.36',
}

import requests
from lxml import etree

html=requests.get(url)
html.encoding=html.apparent_encoding

html_text=html.text
# print(html_text)

selector=etree.HTML(html_text)




title_xpath_str="/html/body/div[@class='container main']/div[@class='row']/div[@class='col middle-column']/div[@class='article']/div[@class='article-body']/div[@id='content']/h1/text()"
titles=selector.xpath(title_xpath_str)
for each in titles:
    print(each.strip())


code_xpath_str="//ol[@class='commentlist']/li[@class='comment even thread-even depth-1']/div[@class='comt-main']/pre/text()"
codes=selector.xpath(code_xpath_str)
for each in codes:
    print(each.strip())

```
