```python
host_url='http://www.qichacha.com/search?key='

import requests,sys,time,re
from lxml import etree
from urllib.request import quote
headers={
'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
'Accept-Encoding':'gzip, deflate',
'Accept-Language':'zh-CN,zh;q=0.9',
'Connection':'keep-alive',

'Host':'www.qichacha.com',
'Referer':'http://www.qichacha.com/user_login',
'Upgrade-Insecure-Requests':'1',
'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36'
}

keyword='阿里云'
url=host_url+quote(keyword)
print(url)
# sys.exit()

response=requests.get(url,headers=headers)
```

```python

html_text=response.text

print(html_text)



```

```python
selector=etree.HTML(html_text)
```

```python
xpath_str="//div[@id='ajaxlist']/section[@id='searchlist']/table[@class='m_srchList']/tbody/tr[1]/td[2]"
links_str="//div[@id='ajaxlist']/section[@id='searchlist']/table[@class='m_srchList']/tbody/tr[1]/td[2]/a[@class='ma_h1']/@href"


content=selector.xpath(xpath_str)
for each in content:
    print(each.xpath('string(.)'))
    
content=selector.xpath(links_str)
for each in content:
    print(each)
```
