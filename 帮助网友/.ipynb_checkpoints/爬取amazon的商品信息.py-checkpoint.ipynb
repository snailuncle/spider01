```python
import requests
```

```python
url='https://www.amazon.com/s/ref=sr_pg_6?fst=as%3Aoff&rh=n%3A3760911%2Cn%3A%2111055981%2Cn%3A11060451%2Cn%3A11060711%2Cn%3A11062031%2Cn%3A11061121%2Cp_72%3A1248876011&page=6&bbn=11061121&ie=UTF8&qid=1522069037&ajr=1'
```

```python
headers={
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding':'gzip, deflate',
    'Accept-Language':'zh-CN,zh;q=0.9',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',
    'Host':'www.amazon.com',
    'Referer':'https://www.amazon.com/',
    'Upgrade-Insecure-Requests':'1',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko),Chrome/65.0.3325.32 Safari/537.36',
    'Content-Type':'text/html; charset=utf-8'
}
```

```python
html=requests.get(url,headers=headers)
```

```python
html_text=html.text
```

```python
from lxml import etree
```

```python
selector=etree.HTML(html_text)
```

```python
xpath_str="//li/div[@class='s-item-container']/div[@class='a-row'][2]/div[@class='a-row a-spacing-none a-spacing-top-mini']/a[@class='a-link-normal s-access-detail-page  s-color-twister-title-link a-text-normal']/h2[contains(@class,'s-access-title')]/text()"
```

```python
content=selector.xpath(xpath_str)
```

```python
for each in content:
    print(each)
```
